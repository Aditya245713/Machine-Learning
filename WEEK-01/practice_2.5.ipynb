### Part A â€“ Central Tendency and Spread
#### 1. Marks: 50, 60, 65, 70, 75, 80, 85, 90
#### Find mean, median, mode.
#### Replace 90 with 900 and recompute mean. Explain the effect of outliers.

```python
import numpy as np
import pandas as pd

marks_list = [50, 60, 65, 70, 75, 80, 85, 90]
marks = pd.DataFrame({'Marks': marks_list})

mean_val = marks['Marks'].mean()
median_val = marks['Marks'].median()
mode_val = marks['Marks'].mode()[0]

print(f"Mean: {mean_val}")
print(f"Median: {median_val}")
print(f"Mode: {mode_val}")
```

```python
marks_list = [50, 60, 65, 70, 75, 80, 85, 900]
marks2 = pd.DataFrame({'Marks': marks_list})
print(f"New Mean: {marks2['Marks'].mean()}")
#prev mean was 71.875, a huge change is seen due to an outlier
```

### Part B â€“ Percentiles, IQR and Z-Score
#### 3. Cat weights (kg): 2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5
#### Find P25, P50, P75 and IQR. Mark possible outliers.

```python
wt_lst = [2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5]

Q1 = np.percentile(wt_lst, 25)
Q2 = np.percentile(wt_lst, 50)
Q3 = np.percentile(wt_lst, 75)
IQR = Q3-Q1

print("P25:", Q1)
print("P50:", Q2)
print("P75:", Q3)
print("IQR:", IQR)

lw_bnd = Q1 - IQR * 1.5
up_bnd = Q3 + IQR * 1.5
print(f"lower bound: {lw_bnd}, upper bound: {up_bnd}")
```
